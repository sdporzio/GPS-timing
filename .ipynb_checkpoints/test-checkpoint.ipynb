```python
reload(io)

run,file = 172,1
mt, df, mdf = io.importFile(run,file)
```

```python
# This first parts wants to create a series of ranges where the LED is off
# Between the end of the pulse (Tt3) and the beginning of the next one (Tl3)
# We also need to be careful about the DAQ occasionally missing triggers

i,j = 0,0 # i for tt3, j for tl3
max_i, max_j = len(df['Tt3']),len(df['Tl3']) # the total number of tt3 and tl3 we have
thresh = 10 # in ms. Sanity check, we need the next trigger to be at least 10 ms after the previous one

t1s,t2s = [],[] # Array of starting/ending windows
# Keep going until you run out of tt3s
# We start with i=0, j=0
while i<max_i:
    # Let's not try to call items we don't have, break before causing errors
    if (i>=max_i) or (j>=max_j):
        break
    # Let's pick up the actual values of tt3 and tl3
    t1 = df['Tt3']['time'].iloc[i]
    t2 =  df['Tl3']['time'].iloc[j]
    # A correct tt3/tl3 pair is a gap in which tl3 comes AFTER tt3 and it's at least 10 ms after
    while (t2<t1):
        # If tl3 is before tt3 it means we need to keep pushing forward
        j+=1
        # Let's not try to call items we don't have, break before causing errors
        if (i>=max_i) or (j>=max_j):
            break
        # Call tl3 to make sure the while loop is happy with it
        t2 =  df['Tl3']['time'].iloc[j]
    # We're out of the loop, it means this tl3 is after tt3, let's make sure it's within threshold
    if (t2>t1) and (t2<(t1+thresh)):
        # All is good, this is a good pair, let's save them and move forward
        t1s.append(t1)
        t2s.append(t2)
    i+=1
```

```python
allPhots = np.array([])
# We have our nice pairs, defining timing ranges after LED, where fluorescence should occur
for i in tqdm(range(len(t1s))):
    # The reference trigger for this period is tt3
    ref_trig = t1s[i]
    # Let's find all the photons between the current tt3 and tl3 defining the timing range
    thesePhots_abs = df['Tl2'].query(f'time>{t1s[i]} & time<{t2s[i]}')['time'].values
    # Now let's find the time since the LED shut down. Let's subtract the reference tt3 trigger
    thesePhots = thesePhots_abs - ref_trig
    # Let's append them to our np.array
    allPhots = np.concatenate((allPhots,thesePhots),axis=None)
```

```python
nbins = 100
xlims = [0,8]

plt.figure(figsize=(5,4),facecolor='white',dpi=100)
hp.HepHist(allPhots,nbins=nbins,xlims=xlims,color=mc[0],yLog=False,xLog=False,label='Corr.')
plt.title('Time since LED switching off')
plt.xlabel('Time [ms]')
plt.xlim(xlims)
plt.ylim(1e1,2000)
plt.show()
```
